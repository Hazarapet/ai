```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 10, 100)

y1 = .5*xs + xs
y2 = -0.4*xs - 3

Y1 = np.random.normal(y1[:50], 4)
Y2 = np.random.normal(y2[50:], 4)
```

```python
plt.plot(xs, y1)
plt.plot(xs, y2)
plt.show()
```

```python
plt.scatter(xs[:50], Y1)
plt.scatter(xs[50:], Y2)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def h(x, theta):
    return sigmoid(x.dot(theta))
```

```python
def loss(x, y, theta):
    n = x.shape[0]
    return - (1 / n) * np.sum(y * np.log(h(x, theta)) + (1 - y) * np.log(1 - h(x, theta)))
```

```python
def gradient(x, y, theta):
    n = x.shape[0]
    grad_theta = np.zeros(theta.shape[0])
    for j in range(theta.shape[0]):
        grad_theta[j] = np.sum((h(x, theta) - y) * x[:, j])
        
    return grad_theta
```

```python
Epoch = 60
alpha = 0.01

real_y = np.concatenate([Y1, Y2])
X = np.column_stack([np.ones(xs.shape[0]), xs, real_y])
Y = np.concatenate([np.ones(Y1.shape[0]), np.zeros(Y2.shape[0])])

theta = np.random.random(X.shape[1])

theta.shape
_loss = []

for epoch in range(Epoch):
    theta = theta - alpha * gradient(X, Y, theta)
    _l = loss(X, Y, theta)
    _loss.append(_l)
    
plt.plot(_loss)
plt.show()
```

```python
print(theta)
dis_bound = -(theta[0] + xs * theta[1]) / theta[2]

plt.plot(xs, dis_bound)
plt.scatter(xs[:50], Y1)
plt.scatter(xs[50:], Y2)
plt.show()
```
