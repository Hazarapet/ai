```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 3
x1, y1 = 2, 4
rand_x1 = np.random.normal(x1, 0.5, size=100)
rand_y1 = np.random.normal(y1, 0.5, size=100)

x2, y2 = 3, 1
rand_x2 = np.random.normal(x2, 0.5, size=100)
rand_y2 = np.random.normal(y2, 0.5, size=100)

x3, y3 = 1, 2
rand_x3 = np.random.normal(x3, 0.5, size=100)
rand_y3 = np.random.normal(y3, 0.5, size=100)

points1 = np.column_stack([rand_x1, rand_y1])
points2 = np.column_stack([rand_x2, rand_y2])
points3 = np.column_stack([rand_x3, rand_y3])

points = np.row_stack([points1, points2, points3])
np.random.shuffle(points)

plt.scatter(rand_x1, rand_y1)
plt.scatter(rand_x2, rand_y2)
plt.scatter(rand_x3, rand_y3)
plt.show()

```

```python
def distance(p1, p2):
    diff = p2 - p1
    diff = diff**2
    return np.sqrt(np.sum(diff))
```

```python
centroids = np.random.random((3, 2)) * 5
```

```python
plt.scatter(points[:, 0], points[:, 1], s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200)
plt.show()
```

```python
for i in range(5):
    c_index = np.zeros(points.shape[0])
    for i, point in enumerate(points):
        _dist = []
        for cent in centroids:
            _dist.append(distance(point, cent))
        _index = np.argmin(_dist) + 1
        c_index[i] = _index
        
    for i in range(k):
        _index = np.argwhere(c_index == i + 1)
        _index = np.squeeze(_index)
        _cluster_points = points[_index]
        plt.scatter(_cluster_points[:, 0], _cluster_points[:, 1])
        centroids[i] = np.average(_cluster_points, axis=0)

    plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200)
    plt.show()
    
```
