```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 10, 100)
ys = xs - 2*np.cos(xs)
shift = np.random.normal(ys, 2)

plt.figure(figsize=(10, 6))
plt.scatter(xs, shift, s=50, color='red')
plt.plot(xs, ys, linewidth=4)
plt.show()
xs.shape
```

```python
def h(theta, x):
    return x.dot(theta)
```

```python
def cost(theta, X, y):
    n = X.shape[0]
    return 1/(2*n)*(X.dot(theta) - y).dot(X.dot(theta) - y)
```

```python
ones = np.ones(xs.shape[0])

X = np.column_stack([ones, xs])
```

```python
theta = [10, -8]

predictions = h(theta, X)
loss = cost(theta, X, shift)

print(loss)
plt.figure(figsize = (10, 6))
plt.scatter(xs, shift, color='green')
plt.plot(xs, predictions, linewidth=4)
plt.show()
```

```python
def closed_form(X, y):
    return np.linalg.pinv(X.T.dot(X)).dot(X.T).dot(y)
```

```python
best_theta = closed_form(X, shift)
best_theta
```

```python
predictions = h(best_theta, X)
loss = cost(best_theta, X, shift)

print(loss)

plt.figure(figsize = (10, 6))
plt.scatter(xs, shift, color='green')
plt.plot(xs, predictions, linewidth=4)
plt.show()
```

```python
xs2 = X[:, 1]

xs2 = xs2*xs2

X2 = np.column_stack([X, xs2])
```

```python
best_theta2 = closed_form(X2, shift)
best_theta2
```

```python
predictions = h(best_theta2, X2)
loss = cost(best_theta2, X2, shift)

print(loss)

plt.figure(figsize = (10, 6))
plt.scatter(xs, shift, color='green')
plt.plot(xs, predictions, linewidth=4)
plt.plot(xs, ys, linewidth=4)
plt.show()
```
